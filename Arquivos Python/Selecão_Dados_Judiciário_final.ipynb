**Normalização do nome das colunas**

```python
dfs = [df_STJ1, df_STJ2, df_STJ3, df_STJ4, df_STJ5]
for i in range(len(dfs)):
    dfs[i].columns = dfs[i].columns.str.strip().str.lower()

df_STJ = pd.concat(dfs, ignore_index=True)
display(df_STJ.head(5))
```

**Normalização da codigo_assuntos e seperação de código por linha sem {}**

```python
if 'codigos_assuntos' in df_STJ.columns:
    df_STJ['codigos_assuntos'] = df_STJ['codigos_assuntos'].astype(str).str.replace(r"[{}]", "", regex=True).str.strip()

    df_STJ['codigos_assuntos'] = df_STJ['codigos_assuntos'].str.split(',')
    df_STJ = df_STJ.explode('codigos_assuntos')

    df_STJ['codigos_assuntos'] = pd.to_numeric(df_STJ['codigos_assuntos'], errors='coerce').astype('Int64')
else:
    print("Column 'codigos_assuntos' not found in DataFrame.")

display(df_STJ.head(5))
```

**Arquivo de assustos e remoção das colunas desnecessárias**

```python
df_assuntos = pd.read_csv("/content/assuntos.csv",  sep=';', on_bad_lines='skip')

display(df_assuntos.head(5))
```

```python
df_assuntos =df_assuntos.drop(columns=['situacao', 'dat_alteracao', 'cod_pai',  'cod_filhos', 'cod_filhos_ativos', 'nivel'])

display(df_assuntos.head(5))
```

**Relação entre codigos_assuntos e codigo e criação da coluna nome_assuntos**

```python
df_STJ['codigos_assuntos'] = pd.to_numeric(df_STJ['codigos_assuntos'], errors='coerce').astype('Int64')

df_assuntos['codigo'] = pd.to_numeric(df_assuntos['codigo'], errors='coerce').astype('Int64')

df_resultado = df_STJ.merge(df_assuntos, left_on='codigos_assuntos', right_on='codigo', how='left')

df_resultado.head(5)
```

**Definição de ano, mes, nome ultima classe, código órgão e outros para inteiros e não float**

```python
df_STJ = pd.concat([ df_resultado], ignore_index=True)

df_STJ['ano'] = df_STJ['ano'].astype(int)
df_STJ['mes'] = df_STJ['mes'].astype(int)
df_STJ['codigo_da_ultima_classe'] = df_STJ['codigo_da_ultima_classe'].astype(int)
df_STJ['codigo_orgao'] = df_STJ['codigo_orgao'].astype(int)

df_STJ.head(5)
```

**Padronização dos dados para sem acentuação e caracter distinto**

```python
import unicodedata

def normalizar_texto(texto):
    if isinstance(texto, str):
        texto = unicodedata.normalize('NFKD', texto).encode('ascii', 'ignore').decode('utf-8')
        texto = texto.replace("?", "A").replace("?", "A")
    return texto

df_STJ = df_STJ.applymap(normalizar_texto)

display(df_STJ.head(5))
```

```python
df_STJ =df_STJ.drop(columns=['codigo'])

df_STJ.rename(columns={'descricao': 'descricao_assunto'}, inplace=True)
```
